```python
result= pd.read_sql_query("""SELECT city, host_location, host_id
                             FROM listings
                             WHERE city = 'Geneva' AND host_location NOT LIKE 'Geneva%'""", connection)
result
```

```python
result= pd.read_sql_query("""SELECT city,
                             SUM(CASE WHEN minimum_nights < 30 THEN 1 ELSE 0 END) AS short_term_listings,
                             SUM(CASE WHEN minimum_nights >= 30 THEN 1 ELSE 0 END) AS long_term_listings
                             FROM listings
                             GROUP BY city""", connection)
result
```

```python
result= pd.read_sql_query("""SELECT l.city, l.id,
                             ROUND(
                             (SUM(CASE WHEN c.available = 't' THEN 1 ELSE 0 END))
                             / COUNT(c.listing_id) * 100,
                             2) AS availability_percentage
                             FROM listings AS l JOIN calender AS c
                             ON l.id = c.listing_id
                             WHERE l.id IN (38425685, 27963222, 393699, 11710930)
                             GROUP BY l.id""", connection)
result
```

```python
result= pd.read_sql_query("""WITH av_table AS (SELECT c.listing_id AS cl, COUNT(available)AS AVL FROM listings AS l JOIN calender AS c
                             ON l.id = c.listing_id
                              WHERE listing_id IN (38425685, 27963222, 393699, 11710930) AND available = 't'
                              GROUP BY c.listing_id)
                              WITH av_table2 AS (SELECT c.listing_id AS cl, COUNT(c.listing_id) AS total_true FROM listings AS l JOIN calender AS c
                             ON l.id = c.listing_id
                              WHERE listing_id IN (38425685, 27963222, 393699, 11710930)
                              GROUP BY c.listing_id)
                              SELECT cl, AVL/total_true*100 AS availability FROM av_table JOIN av_table2 ON av_table.cl = av_table2.cl""", connection)
result
```

```python
result= pd.read_sql_query("""WITH
  av_table AS (
    SELECT
      c.listing_id AS cl,
      COUNT(available) AS AVL
    FROM
      listings AS l
    JOIN
      calender AS c ON l.id = c.listing_id
    WHERE
      listing_id IN (38425685, 27963222, 393699, 11710930) AND available = 't'
    GROUP BY
      c.listing_id
  ),
  av_table2 AS (
    SELECT
      c.listing_id AS cl2,
      COUNT(c.listing_id) AS total_true
    FROM
      listings AS l
    JOIN
      calender AS c ON l.id = c.listing_id
    WHERE
      listing_id IN (38425685, 27963222, 393699, 11710930)
    GROUP BY
      c.listing_id
  )
SELECT
  av_table.cl,
  AVL / total_true * 100 AS availability
FROM
  av_table
JOIN
  av_table2 ON av_table.cl = av_table2.cl2
  GROUP BY av_table.cl

""", connection)
```
